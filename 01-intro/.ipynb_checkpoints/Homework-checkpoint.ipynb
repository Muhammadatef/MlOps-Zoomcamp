```python
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet')
```

#### Q2. Computing duration

#### Now let's compute the duration variable. It should contain the duration of a ride in minutes.

#### What's the standard deviation of the trips duration in January?

```python
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
```

```python
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
```

```python
df
```

```python
df
```

```python
td = df.duration.iloc[0]
td.total_seconds() / 60
```

```python
df.duration = df.duration.apply(lambda td : td.total_seconds() / 60)
```

```python
df.duration.describe()
```

```python
df['duration']
```

### Q3. Dropping outliers

#### Next, we need to check the distribution of the duration variable. There are some outliers. Let's remove them and keep only the records where the duration was between 1 and 60 minutes (inclusive).



```python
((df.duration >= 1) & (df.duration <= 60)).mean()
```

```python
df = df[(df.duration >= 1) & (df.duration <= 60)]
```

```python
df.duration.describe(percentiles=[0.95, 0.98,0.99])
```

```python
df
```

#### What fraction of the records left after you dropped the outliers?

```python
3009173 /3066766 
```

```python
categorical = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']
```

```python
df[categorical] = df[categorical].astype(str)
```

```python
df['PULocationID'] = df['PULocationID'].astype(str)
df['DOLocationID'] = df['DOLocationID'].astype(str)
```

```python
df.dtypes
```

```python
train_dicts = df[categorical + numerical].to_dict(orient='records')
```

```python
train_dicts
```

```python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
```

```python
target = 'duration'
y_train = df[target].values
```

```python
y_train
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
```

```python
sns.distplot(y_pred, label='prediction')
sns.distplot(y_train, label='actual')


plt.legend()
```

### Q6 Evaluating the model

#### Now let's apply this model to the validation dataset (February 2023).

#### What's the RMSE on validation?


```python
# Assume val_df is your validation DataFrame for February 2023

val_df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet')
```

```python
val_df.tpep_dropoff_datetime = pd.to_datetime(val_df.tpep_dropoff_datetime)
val_df.tpep_pickup_datetime = pd.to_datetime(val_df.tpep_pickup_datetime)
val_df['duration'] = val_df.tpep_dropoff_datetime - val_df.tpep_pickup_datetime
```

```python
# 1. Prepare features
val_df['PULocationID'] = val_df['PULocationID'].astype(str)
val_df['DOLocationID'] = val_df['DOLocationID'].astype(str)
```

```python
val_dicts = val_df[['PULocationID', 'DOLocationID']].to_dict(orient='records')
```

```python
# 2. Transform validation features using training DictVectorizer
X_val = dv.transform(val_dicts)
```

```python
# 3. Prepare validation target in minutes
val_df['duration'] = val_df['duration'].dt.total_seconds() / 60
y_val = val_df['duration'].values
```

```python
# 4. Predict
y_pred_val = lr.predict(X_val)
```

```python
# 5. Calculate RMSE
from sklearn.metrics import mean_squared_error
import numpy as np

mse_val = mean_squared_error(y_val, y_pred_val)
rmse_val = np.sqrt(mse_val)
print(round(rmse_val, 2))
```
